```python
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
utils = importr('utils')

# Set a CRAN mirror before installing packages
ro.r('options(repos = c(CRAN = "https://cloud.r-project.org"))')

# Install devtools if not already installed
utils.install_packages('devtools')

# Install toRvik from GitHub
ro.r('devtools::install_github("andreweatherman/toRvik")')

# Now try loading it
ro.r('library(toRvik)')

# Example: Call an R function from the package
result = ro.r('bart_players(year = 2024)')
result
```
